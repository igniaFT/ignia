```python
import math

def giai_phuong_trinh_bac_hai(a, b, c):
    # Kiểm tra nếu a = 0 (không phải phương trình bậc hai)
    if a == 0:
        if b == 0:
            if c == 0:
                return "Phương trình có vô số nghiệm (0x + 0 = 0)."
            else:
                return "Phương trình vô nghiệm (0x + c = 0)."
        else:
            x = -c / b
            return f"Đây là phương trình bậc nhất. Nghiệm duy nhất là x = {x:.2f}"
    
    # Tính delta (biệt thức)
    delta = b**2 - 4*a*c
    
    if delta > 0:
        # Hai nghiệm phân biệt
        x1 = (-b + math.sqrt(delta)) / (2*a)
        x2 = (-b - math.sqrt(delta)) / (2*a)
        return f"Phương trình có hai nghiệm phân biệt:\nx1 = {x1:.2f}\nx2 = {x2:.2f}"
    elif delta == 0:
        # Nghiệm kép
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép:\nx1 = x2 = {x:.2f}"
    else:
        # Nghiệm phức
        phan_thuc = -b / (2*a)
        phan_ao = math.sqrt(abs(delta)) / (2*a)
        return f"Phương trình có hai nghiệm phức:\nx1 = {phan_thuc:.2f} + {phan_ao:.2f}i\nx2 = {phan_thuc:.2f} - {phan_ao:.2f}i"

# --- Phần nhập liệu từ người dùng ---
try:
    print("Nhập các hệ số cho phương trình ax^2 + bx + c = 0:")
    # Chuyển đổi giá trị nhập vào thành số thực (float)
    a = float(input("Nhập a: "))
    b = float(input("Nhập b: "))
    c = float(input("Nhập c: "))
    
    # Gọi hàm giải phương trình và in kết quả
    ket_qua = giai_phuong_trinh_bac_hai(a, b, c)
    print("\n--- Kết quả ---")
    print(ket_qua)
    
except ValueError:
    print("Lỗi: Vui lòng nhập các giá trị số hợp lệ.")

```
